```python
%matplotlib notebook
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

it = 1000

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot([0, 1], [0, 1], linestyle="-", markersize=1, color="red")
line_f, = ax.plot([ i/it for i in range(it)], [0]*it, linestyle="-", marker="", markersize=1, color="blue")
line_x, = ax.plot([0]*it, [0]*it, linestyle="-", marker="o", markersize=1, color=(0,1,0,0.5))
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])


def update(x0=0.5, r=0.1):
    f = []
    for i in range(it):
        i = i/it
        f.append(r*i*(1-i))

    x, y = [x0], [0]
    for i in range(it-1):
        if i % 2 == 0:
            x.append(x[-1])
            y.append(r*x[-1]*(1-x[-1]))
        else:
            x.append(y[-1])
            y.append(y[-1])

    line_f.set_ydata(f)
    line_x.set_xdata(x)
    line_x.set_ydata(y)
    fig.canvas.draw()
    
interact(update, x0=(0.0,1.0,0.01), r=(0.0,4.0,0.01))
```

```python
%matplotlib notebook
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

it = 200

fig, ax = plt.subplots()
fig.set_tight_layout(True)

lines = [
    ax.plot([j/it for j in range(it)], [i/it]*it, linestyle="", marker="o", markersize=1, color=(0, 0, 1, i/it))[0]
    for i in range(it)
]

ax.set_xlim([0, 1])
ax.set_ylim([0, 1])


def update(r=0.1):
    
    y = [
        j/it for j in range(it)
    ]

    for i in range(it):
        for j in range(it):
            y[j] = r*y[j]*(1-y[j])
        
        lines[i].set_ydata(list(y))
            
    fig.canvas.draw()
    
interact(update, r=(0.0,4.0,0.01))
```

```python
%matplotlib notebook
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

# mapas de lorenz

it = 10000
dt = 0.01

fig, ax = plt.subplots()
fig.set_tight_layout(True)

a = 0.5
b = 0.5

x = [0.5]
y = [0.5]

for i in range(it-1):
    x.append(y[i-1] - a*x[i-1]**2)
    y.append(b*x[i-1])

line, = ax.plot(x, y, linestyle="-", marker="o", markersize=1, color=(0, 0, 1))

ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])


def update(x0=0.5, y0=-0.5, a=0.5, b=-0.5):

    x = [x0]
    y = [y0]

    for i in range(it-1):
        x.append(y[i-1]*dt - a*x[i-1]**2*dt)
        y.append(b*x[i-1]*dt)
        
    line.set_xdata(x)
    line.set_ydata(y)
            
    fig.canvas.draw()
    
interact(update, x0=(-1.0,1.0,0.01), y0=(-1.0,1.0,0.01), a=(-1.0,1.0,0.01), b=(-1.0,1.0,0.01))
```
