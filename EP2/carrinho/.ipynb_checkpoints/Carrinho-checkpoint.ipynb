```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import glob
names = []
for file in glob.glob("*.csv"):   #armazena o titulo de cada arquivo em uma lista
    names.append(file)
names.sort()
i = 0
for i in range(len(names)):
    tbl = pd.read_csv(names[i])
    tbl.head()
    fig = plt.figure(figsize=(32,4))
    plt.subplot(131)
    plt.grid(True)
    plt.xlabel("Time")
    plt.ylabel("m/s^2")
    plt.plot(tbl["time"], tbl["ax"], label = "X axis")
    plt.plot(tbl["time"], tbl["ay"], label ="Y axis")
    plt.plot(tbl["time"], tbl["az"], label ="Z axis")
    plt.plot(tbl["time"], tbl["aT"], label = "Total acceleration")
    plt.title(names[i])
    plt.legend()
    plt.show()
```
