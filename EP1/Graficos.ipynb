```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import glob
names = []
for file in glob.glob("*.csv"):   #armazena o titulo de cada arquivo em uma lista
    names.append(file)
#print(names)
names.sort()  #ordena para facilitar a interpretacao

def calcula(i):
    #funcao que gera cada grafico
    tbl = pd.read_csv(names[i])  #le os arquivos.csv
    tbl.head() 
    fig = plt.figure(figsize=(14, 3.5))  #formata os graficos
    plt.subplot(131+i%3)  #posicao de cada grafico na linha (isso que nao consegui fazer funcionar nesse codigo)
    plt.xlabel('Time')
    plt.ylabel("FN/Fg")
    plt.plot(tbl['time'],tbl['gFx'],label="gFx") 
    plt.plot(tbl['time'],tbl['gFy'],label="gFy")
    plt.plot(tbl['time'],tbl['gFz'],label="gFz")   #cria cada linha
    #sns.regplot(tbl['time'],tbl['gFz'])
    plt.title(names[i])
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()     #printa o grafico
```

```python
for i in range(len(names)):
    calcula(i)
#plt.tight_layout()
#plt.show()
```
