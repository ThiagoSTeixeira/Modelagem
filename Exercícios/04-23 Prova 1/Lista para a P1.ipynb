```python
# Implementa o exercício 1 

import math
import matplotlib.pyplot as pyplot

g = 9.8

def nextYeuler(y, v, dt):
    """ nextYeuler implementa o método de Euler. 
        
    Parâmetros:
    -----------
    y: float - Posição atual
    v: float - Velocidade atual
    dt: float - passo de tempo
    
    Returns:
    --------
    y: float : posição atualizada
    
    """
    ### BEGIN SOLUTION
    y = y + v*dt
    return y
    ### END SOLUTION
    
def nextVeuler(v,dt):
    """
    Parâmetros:
    ------------
    v: velocidade atual
    dt: passo de tempo 
    
    Returns: 
    ---------
    v: float: velocidade atualizada 
    
    """
    return v - g * dt
    

def nextYana(t, yi, vi):
    """ nextYana implementa o método analítico. 

    Parâmetros:
    -----------
    y: float - Posição atual
    yi: float - Posição inicial
    vi: float - Velocidade inicial    
    t: float - Tempo atual
    
    Returns:
    --------
    x: float : posição atualizada
    
    """
    y = yi + vi * t - ((g* t* t)/2)
    return y
  
def nextVana(t, vi):
    """ nextXana implementa o método analítico. 

    Parâmetros:
    -----------
    vi: float - Velocidade inicial    
    t: float - Tempo atual
    
    Returns:
    --------
    v: float : velocidade atualizada
    
    """
    v = vi - g*t
    return v

def main():
    # tempo inicial
    t= 0.0
    # tempo final
    tf= 3.0
    # passo de tempo
    dt= 0.1
    # posição inicial
    yi= 0.0
    # velocidade inicial    
    vi= 0.0
    # posição atual
    y= 0.0
    # velocidade atual    
    v= 0.0
    # vetor para guardar os tempos do movimento
    vt=[t]
    # Forma de criar um vetor vazio em Python
    # vetor para guardar as posições da partícula segundo a forma analítica 
    vya=[]
    # vetor para guardar as posições da partícula segundo Euler    
    vye=[]
    # vetor para guardar as velocidades da partícula segundo a forma analítica 
    vva=[]
    # vetor para guardar as velocidades da partícula segundo Euler 
    vve=[] 
    
    ### BEGIN SOLUTION

    while (t<tf):
        vya.append(nextYana(t,yi,vi)) # adiciona a posição analítica
        vva.append(nextVana(t, vi))   # adiciona a velocidade analítica     
        vye.append(y)                 # adiciona a posição por euler
        vve.append(v)                 # adiciona a velocidade por euler
        print('Position error ' + str(abs(y - nextYana(t,yi,vi))))  # erro da posição 
        print('Velocity error ' + str(abs(v - nextVana(t, vi))))    # erro da velocidade
        t += dt                       # atualiza o tempo
        vt.append(t)                  
        y = nextYeuler(y, v, dt)
        v = nextVeuler(v, dt)
        
    ### END SOLUTION
    
    pyplot.figure(0)
    pyplot.plot(vt[0:-1],vye,label='Euler',linestyle='',marker='o') 
    pyplot.plot(vt[0:-1],vya,label='Analytical') 
    pyplot.title('Position')
    pyplot.show(block=False)
    
    pyplot.figure(1)
    pyplot.plot(vt[0:-1],vve,label='Euler',linestyle='',marker='o') 
    pyplot.plot(vt[0:-1],vva,label='Analytical') 
    pyplot.title('Velocity')
    pyplot.show(block=False)

main() 
```

Ex7. (b) A forma analítica de v(t) é $v(t) = -y(0) \cdot \omega_0\cdot \sin(\omega_0 * t) + v(0)\cdot \cos(\omega_0 *t)$. Vamos mostrar que (2.12) satisfaz (2.11):

$\frac{d}{dt}\Big[ y(0)\cdot \cos(\omega_0\cdot t) + \frac{v(0)}{\omega_0} \cdot \sin(\omega_0 \cdot t)\Big] \\$ 

$= -y(0)\cdot \sin(\omega_0\cdot t)\cdot\omega_0  + \frac{v(0)}{\omega_0} \cdot \cos(\omega_0 \cdot t)\cdot \omega_0 $

$= -y(0)\cdot \sin(\omega_0\cdot t)\cdot\omega_0  + v(0) \cdot \cos(\omega_0 \cdot t)  = v$

$ \implies \frac{dy}{dt} = v $

Agora vamos mostrar que $\frac{dv}{dt} = -\frac{k}{m}\cdot x$. Sabemos $\omega_0 = \sqrt{\frac{k}{m}}$

$= \frac{d}{dt}\Big[-y(0)\cdot \sin(\omega_0\cdot t)\cdot\omega_0  + v(0) \cdot \cos(\omega_0 \cdot t) \Big]$

$= -y(0)\cdot \cos(\omega_0\cdot t)\cdot\omega_0^2  - v(0) \cdot \sin(\omega_0 \cdot t)\cdot \omega_0 $

$= -\omega_0^2\Big[ y(0)\cdot \cos(\omega_0\cdot t)+  \frac{v(0)}{\omega_0} \cdot \sin(\omega_0 \cdot t) \Big]$

$= -\omega_0^2 \cdot x$

$= -\frac{k}{m} \cdot x $

$\implies \frac{dv}{dt} = -\frac{k}{m} x$

```python
# Implementa o exercício 8

import math
import matplotlib.pyplot as pyplot

g = 9.8

def nextYeuler(y, v, dt):
    return y + v*dt
    
def nextVeuler(v,dt):
    return v - g * dt
  
def nextYeulerR(y, v, dt):
    return y + v*dt/2
    
def nextVeulerR(v,dt):
    return v - g * dt/2
    
def nextYana(t, yi, vi):
    return yi + vi * t - ((g* t* t)/2)
  
def nextVana(t, vi):
    return vi - g*t

def main():
    # tempo inicial
    t= 0.0
    # tempo final
    tf= 3.0
    # passo de tempo
    dt= 0.01
    # posição inicial
    yi= 10.0
    # velocidade inicial    
    vi= 0.0
    # posição atual
    y= yi
    # velocidade atual    
    v= 0.0
    # vetor para guardar os tempos do movimento
    vt=[t]
    # Forma de criar um vetor vazio em Python
    # vetor para guardar as posições da partícula segundo a forma analítica
    vya=[]
    # vetor para guardar as posições da partícula segundo Euler    
    vye=[]
    # vetor para guardar as velocidades da partícula segundo a forma analítica 
    vva=[]
    # vetor para guardar as velocidades da partícula segundo Euler 
    vve=[] 
    
    ### BEGIN SOLUTION

    while (y >=0 ):
        vya.append(nextYana(t,yi,vi)) # adiciona a posição analítica
        vva.append(nextVana(t, vi))   # adiciona a velocidade analítica     
        vye.append(y)                 # adiciona a posição por euler
        vve.append(v)         
        #print('VELOCITY: Euler-Richardson: {0:0.6f}'.format(abs(v - nextVana(t, vi))))
        t += dt                       # atualiza o tempo
        vt.append(t)        
        v = nextVeuler(v, dt)  
        y = nextYeuler(y, v, dt)
        
    print('POSITION: Euler-Richardson: {0:0.6f}'.format(abs(y - nextYana(t,yi,vi))))
    
    pyplot.figure(0)
    pyplot.plot(vt[0:-1],vye,label='Euler',linestyle='',marker='o') 
    pyplot.plot(vt[0:-1],vya,label='Analytical') 
    pyplot.title('Position - Euler')
    pyplot.show(block=False)
    
    pyplot.figure(1)
    pyplot.plot(vt[0:-1],vve,label='Euler',linestyle='',marker='o') 
    pyplot.plot(vt[0:-1],vva,label='Analytical') 
    pyplot.title('Velocity - Euler')
    pyplot.show(block=False)
    

main() 
```

```python
import math
import matplotlib.pyplot as pyplot

def f(a, b, u):
  return a * math.exp(-b * u * u)

def df(a, b, u, dx):
  return (f(a, b, u + dx) - f(a, b, u))/dx 

def main():
  x = -5
  xf = 5
  dx = 0.1
  a = 1
  vx=[x]
  vf1=[]
  vf2=[]
  df1=[]
  df2=[]
  while(x<=xf):
    vf1.append(f(a, 1, x))
    vf2.append(f(a, 4, x))
    df1.append(df(a, 1, x, dx))
    df2.append(df(a, 4, x, dx))
    x += dx
    vx.append(x)
  pyplot.figure(0)
  pyplot.plot(vx[0:-1],vf1,label='f(u), B = 1') 
  pyplot.plot(vx[0:-1],vf2,label='f(u), B = 4') 
  pyplot.plot(vx[0:-1],df1,label='df(u)/du, B = 1') 
  pyplot.plot(vx[0:-1],df2,label='df(u)/dx, B = 4')
  pyplot.title('f(u)')
  pyplot.show(block=False)
    
main()

  
```
