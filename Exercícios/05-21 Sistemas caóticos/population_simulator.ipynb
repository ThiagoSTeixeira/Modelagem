```python
%matplotlib notebook
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

it = 1000

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot([0, 1], [0, 1], linestyle="-", markersize=1, color="red")
line_f, = ax.plot([ i/it for i in range(it)], [0]*it, linestyle="-", marker="", markersize=1, color="blue")
line_x, = ax.plot([0]*it, [0]*it, linestyle="-", marker="o", markersize=1, color=(0,1,0,0.5))
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])


def update(x0=0.5, r=3.63):
    f = []
    for i in range(it):
        i = i/it
        f.append(r*i*(1-i))

    x, y = [x0], [0]
    for i in range(it-1):
        if i % 2 == 0:
            x.append(x[-1])
            y.append(r*x[-1]*(1-x[-1]))
        else:
            x.append(y[-1])
            y.append(y[-1])

    line_f.set_ydata(f)
    line_x.set_xdata(x)
    line_x.set_ydata(y)
    fig.canvas.draw()
    
interact(update, x0=(0.0,1.0,0.01), r=(0.0,4.0,0.01))
```

```python
%matplotlib notebook
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

it = 1000000
dt = 0.01
fig, ax = plt.subplots()
fig.set_tight_layout(True)

x = [0] * it
y = [0] * it

line, = ax.plot(x, y, linestyle="", marker="o", markersize=0.1, color=(0, 0, 1, 0.5))

ax.set_xlim([0, 5])
ax.set_ylim([0, 5])

def update(x0=0.5, y0=0.5, a=0.5, b=0.5, c=0.5, d=0.5):
    x[0] = x0
    y[0] = y0
    

    for i in range(1,it):
        x[i] = x[i-1] + (x[i-1]*(a - b*y[i-1]))*dt
        y[i] = y[i-1] + (y[i-1]*(c*x[i-1] - d))*dt
        
    line.set_xdata(x)
    line.set_ydata(y)
            
    fig.canvas.draw()
    
interact(
    update, 
    x0=(0,1.0,0.1),
    y0=(0,1.0,0.1),
    a=(0,4.0,0.1),
    b=(0,4.0,0.1),
    c=(0,4.0,0.1),
    d=(0,4.0,0.1)
)
```
