```python
from math import ceil

dt = 0.001

max_time = ceil(fall_time / dt) # 1s

t = [i*dt for i in range(0, max_time)]
y = [0] * max_time
v = [0] * max_time
g = -10

y[0] = fall_height
v[0] = 0

for i in range(1, max_time):
    y[i] = y[i-1] + v[i-1]*dt
    v[i] = v[i-1] + g     *dt
    
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
    
ax1.plot(t, y)
ax1.set_xlabel("t (s)")
ax1.set_ylabel("h(t) (m)")

ax2.plot(t, v)
ax2.set_xlabel("t (s)")
ax2.set_ylabel("v(t) (m)")

fig = plt.gcf()
fig.set_size_inches(12, 6)
fig.tight_layout()

plt.show()
```
